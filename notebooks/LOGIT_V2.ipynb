```python
import statsmodels.api as sm

# Fonction pour la slection stepwise forward
def forward_selection(X, y, significance_level=0.05):
    initial_features = X.columns.tolist()
    best_features = []
    
    while len(initial_features) > 0:
        remaining_features = list(set(initial_features) - set(best_features))
        new_pval = pd.Series(index=remaining_features, dtype=float)
        
        for new_column in remaining_features:
            model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[best_features + [new_column]]))).fit()
            new_pval[new_column] = model.pvalues[new_column]
        
        min_p_value = new_pval.min()
        if min_p_value < significance_level:
            best_features.append(new_pval.idxmin())
        else:
            break
    
    return best_features
X = data.drop(["TARGET", "date_mensuelle"], axis=1)
y=data["TARGET"]
selected_features = forward_selection(X, y)
```

```python
from scipy.stats import chi2_contingency
vars_signif = []
for column in df[selected_features]:
    # Construire la table de contingence pour la variable actuelle et la variable cible
    contingency_table = pd.crosstab(df[column], df['TARGET'])
    
    # Effectuer le test du Chi2
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    if p <= 0.01: vars_signif.append(column)
    
```
